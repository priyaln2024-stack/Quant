```python
import yfinance as yf
import matplotlib.pyplot as plt

close = yf.download("SPY", start = "2000-01-01")["Close"]

return_12m = close/close.shift(252)-1
signal = (return_12m>0).astype(int)

returns = close.pct_change()
strat = signal.shift(1)*returns

buy_hold = (1+returns).cumprod()

momentum = (1+strat).cumprod()

plt.plot(buy_hold,label = "Buy & Hold", color = 'Blue')
plt.plot(momentum,label = "Momentum", color = 'Crimson')
plt.legend()
```
